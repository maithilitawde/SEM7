```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
df.head()
```

```python
df.drop(['CustomerId' , 'Surname'] , axis = 1 , inplace = True)
```

```python
df.head()
```

```python
x = df.drop(['Exited' , 'RowNumber'] , axis = 1)
y = df['Exited']
```

```python

x
```

```python
y
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
x_transformer = LabelEncoder()
x['Gender'] = x_transformer.fit_transform(x['Gender'])
x['Geography'] = x_transformer.fit_transform(x['Geography'])
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test ,y_train ,  y_test = train_test_split(x , y , test_size= 0.2 , random_state= 0 )
```

```python
pip install tensorflow
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(units = 6 , activation= 'relu' , input_shape = (x_train.shape[1], )))
```

```python
classifier.add((Dense(units = 6 , activation = 'relu')))
```

```python
classifier.add(Dense(units = 1 , activation = 'sigmoid'))
```

```python
classifier.summary()
```

```python
classifier.compile(optimizer = 'adam' , loss = 'binary_crossentropy' , metrics = ['accuracy'] )
```

```python
classifier.fit(x_train , y_train , validation_split = 0.1 , batch_size = 32 , epochs = 200)
```

```python
prediction = classifier.predict(x_test)
prediction = (prediction>0.5).astype(int) //remember
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix
accuracy_score(y_test , prediction)
```

```python
cm = confusion_matrix(prediction , y_test)
cm
```
