```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import sympy as sym
```

```python
def objective(x):
    return (x+3)**2
```

```python
def derivative(x):
    return 2*(x+3)
```

```python
def gradient(start , alpha , max_iter):
    x_list = []
    x = start
    x_list.append(x)
    precision = 0.000001
    prev_step_size = 1
    iter = 0
    while prev_step_size > precision and iter< max_iter:
        prev_x = x
        gradi = derivative(x)
        x = x - (alpha*gradi)   #step towards neg gradient
        x_list.append(x)   #remember this
        prev_step_size = abs(prev_x - x)
        iter += 1
    return x_list
```

```python
start = 2
alpha = 0.1
max_iter = 1000
```

```python
x_corr = np.linspace(-15 , 15, 100)
plt.plot(x_corr , objective(x_corr))
plt.plot(2 , objective(2) , 'ro')
```

```python
x  = gradient(start , alpha , max_iter)
x_corr = np.linspace(-5 , 5 ,100)
plt.plot(x_corr , objective(x_corr))

x_arr = np.array(x)
plt.plot(x_arr , objective(x_arr) , '.-' , color = 'red' )
plt.show()
```
